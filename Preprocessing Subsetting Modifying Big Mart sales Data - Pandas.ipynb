#### `DIFFERENCE BETWEEN LOC AND ILOC`

---

***Let's create a sample dataframe***

---

```python
sample_df = pd.DataFrame({
    'gender' : ['M', 'F', 'M', 'M', 'F'],
    'grade'  : ['A', 'A', 'B', 'B', 'A'],
    'marks'  : [ 22,  21,  12,  14,  20],
    'id'     : ['A101', 'A102', 'A103', 'A104', 'A105']
})
sample_df
```

---

***Now, when we define a dataframe by default the index is a range of numbers. Let's see what happens if try to slice the dataframe using both `loc` and `iloc`***

---

```python
sample_df.loc[2:4]
```

```python
sample_df.iloc[2:4]
```

---

- When we try to slice the dataframe using the `loc` function on `range(2 to 4)` it first finds out the index with a `label 2` and goes till it finds the index with a `label 4`. 
- On the other hand, when we try to silce the dataframe using the `iloc` function on `range(2 to 4)` it starts with the `starting point index 2` and goes till `end point - 1 which is 3`.

---


***Let's verify this by sorting the dataframe by marks column.***

---

```python
sample_df = sample_df.sort_values(by=['marks'])
```

```python
sample_df
```

---

***Now that we have sorted the dataframes, the indices are shuffled accordingly. Let's try to slice the dataframes again.***

---

```python
sample_df.loc[2:4]
```

```python
sample_df.iloc[2:4]
```

----

- You can see that when we tried to slice the dataframe using the `loc` function it finds out the name of start and end point and slice the data whereas `iloc` still gives the 2nd and 3rd value of the dataframe.

----

***Let's change the index of the dataframe to a categorical variable.***

---

```python
sample_df.set_index('id', drop=True, inplace=True)
```

```python
sample_df
```

---

***Now, try to slice the dataframe using the loc and iloc function.***

---

```python
sample_df.iloc[2:4]
```

```python
sample_df.loc[2:4]
```

***Now, we got an error because no index was with the name 2 and 4 in the dataframe.***

---

#### `HOW TO SELECT ROWS BASED ON CONDITION?`

- Select all rows where the value of `Outlet_Establishment_Year` is 1987

---

```python
#filter row with condition
data.loc[data['Outlet_Establishment_Year'] == 1987]
```

---

#### `HOW TO SELECT ROWS BASED ON MULTIPLE CONDITIONS?`

- When passing multiple conditions make sure that you put each of the condition in a parenthesis ().
- Select all rows where the value of `Outlet_Establishment_Year` is 2009 and value of `Outlet_Size` is Medium.

---

```python
# filter data based on multiple conditions
data.loc[(data.Outlet_Establishment_Year==2009) & (data.Outlet_Size=='Medium')]
```

***If you want to filter for a list of values from a column then instead of writing multiple conditions use the `isin` function.***

---
#### `HOW TO FILTER FOR A LIST OF VALUES?`
---

```python
# get rows for 3 years 1987, 1988, 1999
data[(data.Outlet_Establishment_Year == 1987) | (data.Outlet_Establishment_Year == 1988) | (data.Outlet_Establishment_Year == 1999)]
```

```python
# filter for a list of values, Instead of using multiple condition we can use this
data.loc[data.Outlet_Establishment_Year.isin([1987, 1988, 1999])]
```

---

 - `df.loc[]` provides simpler syntax over `df[]`

#### `HOW TO IMPUTE THE MISSING VALUES USING LOC IN ANY COLUMN?`

---

```python
#check for the null values
data.isna().sum()
```

So, we have missing values in the column `Item_Weight` and `Outlet_Size`.

#### `Select the columns where Item_Weight has missing values using loc`

---

```python
data.loc[data.Item_Weight.isna()==True]
```

```python
# fill the null values in Item_Weight by mean
data.loc[(data.Item_Weight.isna()==True), 'Item_Weight'] = data.Item_Weight.mean()
```

#### `Fill the missing values in the column: Outlet_Size by most frequent value using loc`

***Mode of Outlet Size***

```python
data.Outlet_Size.mode()
```

```python
data.Outlet_Size.value_counts()
```

```python
data.loc[(data.Outlet_Size.isna()==True), 'Outlet_Size'] = 'Medium'
```

```python
data.isna().sum()
```

#### `Use the fillna function to impute the missing values.`

- `fillna` function is another way to impute the missing values. Use the parameter `inplace=True` to store the results in the dataframe.

```python
data.Outlet_Size.fillna('Medium', inplace=True)
```

```python
data.isna().sum()
```

***Let's have a look at the count of each category in the column `Item_Fat_Content`. We will use `value_counts` function to to do that.***


---

```python
data.Item_Fat_Content.value_counts()
```

---

***We can see that the categories `Low Fat`, `LF` and `low fat` are same and also `Regular`, `reg` are same. So to keep the data clean we will map all these to only two categories to `LF` and `R` using the map function. Let's see how?***

```python
# Create a new mapping (dictionary)
mapping = {
    'Low Fat' : 'LF',
    'Regular' : 'R',
    'low fat' : 'LF',
    'reg' : 'R',
    'LF' : 'LF'
}
```

```python
data.Item_Fat_Content.map(mapping)
```

```python
# use the  map function to update the values
data.Item_Fat_Content = data.Item_Fat_Content.map(mapping)
```

```python
# Count of new categories in the column Item_Fat_Content
data.Item_Fat_Content.value_counts()
```
